### What is/are the main feature(s) of interest in your dataset?

I am most interested in finding out which factors can be most useful in predicting the outcome of a loan, mainly, whether it will be successfully repaid or charged-off/defaulted

### What features in the dataset do you think will help support your investigation into your feature(s) of interest?

The feature that indicates the outcome of a loan is contained in the LoanStatus variable. Based on the above, we see that there are current loans, including those that are in good standing as well as delinquent loans. Since we are interested in loans that have had its outcome determined already, we will only look at loans that are marked "Chargedoff", "Completed" and "Defaulted" (for this analysis, we will treat Chargedoff loans as being the final outcome - however, in real life charged off loans do have a small probability of being recovered). I will ignore "Cancelled" loans as it is not clear from the information why the loan has been cancelled and the number of such loans are very small (5).

As the number of variables is very high and there are a significant number of variables which most likely are highly correlated to each other (for example, OpenRevolvingAccounts with OpenRevolvingMonthlyPayment, CreditScoreRangeLower with CreditScoreRangeUpper, etc.), a few variables that are deemed to best represent their category of data should be selected (or engineered) for the exploration phase.

I expect that some of the factors that could potentially indicate the probability of default that would be worth exploring includes, along with an initial hypothesis:
- Term of the loan (longer loans more likely to default?)
- Rate of the loan (higher rate more likely to default?)
- Amount of the loan (higher amount more likely to default?)
- Purposes of loan (loans for "neccessities" such as housing loans less likely to default?)
- Employment status (employed borrowers less likely to default?)
- Recommendations (more recommendations means less likely to default?)
- Credit scores (higher credit score = less likely todefault?)
- Total credit amount (more credit amount = less likely to default?)
- Debt-to-income (less debt to income = less likely to default?)
- Investment from friends amount (more friend investments = less likely to default?)

As I am only interested in loans which have had their final status determined, the data will be filtered for completed loans only to reduce execution time during exploration.

Additionally, in order to limit potential effects due to changes in the credit environment in the USA, particularly due to the Subprime Crisis of 2008, I will use only data from 2011 onwards, by which time the credit environment should have stabilised.

```python
df_clean = df.copy()

df_clean = df_clean.query('LoanStatus == "Chargedoff" | LoanStatus == "Completed" | LoanStatus == "Defaulted"')

df_clean = df_clean.query('ClosedDate > "2011-01-01"')

df_clean
```

This leaves 29,808 rows.

## Univariate Exploration

#### Loan outcomes

Firstly, we take a quick look at the distribution of the loan outcomes: 

```python
fig = plt.figure()
ax = plt.subplot(1, 1, 1)
sns.countplot(x=df_clean['LoanStatus'], palette=sns.color_palette(n_colors=1))
ax.set_title('Distribution of Loan Outcomes')
plt.show()
```

We see that the majority of loans are successfully completed. The remaining are either considered charged-off or defaulted.

In order to simplify the analysis, Chargedoff and Defaulted loans could be considered to be the same outcome:

```python
# Create new variable for loan outcome with binary outcomes (complete or incomplete)
for index, row in df_clean.iterrows():
    if row['LoanStatus'] == 'Completed':
        df_clean.loc[index, 'LoanOutcome'] = 'Completed'
    else:
        df_clean.loc[index, 'LoanOutcome'] = 'Fail-to-collect'
```

```python
fig = plt.figure()
ax = plt.subplot(1, 1, 1)
sns.countplot(x=df_clean['LoanOutcome'], palette=sns.color_palette(n_colors=1),
              order=['Completed', 'Fail-to-collect'])
ax.set_title('Distribution of Loan Outcomes')
plt.show()
```

#### Loan terms

Next, we look at the distribution of loan terms:

```python
fig = plt.figure()
ax = plt.subplot(1, 1, 1)
sns.countplot(x=df_clean['Term'], palette=sns.color_palette(n_colors=1))
ax.set_title('Distribution of Loan Terms')
ax.set_xlabel('Term (months)')
plt.show()
```

We can see that the vast majority of loans are 36-month terms, with the remaining being 60- and 12-month loans. My expectation is that longer loans will be more likely to default, which will be explored in the next section.

#### Loan amounts

Next, we look at the distribution of loan amounts. Since not all loans have the same term, I propose normalising loan amount according to its loan term in order to compare loans with different terms on a similar basis. (There is a MonthlyLoanPayment variable available but it gives 0 for a lot of shorter-term loans)

```python
# Generate loan amount over month by dividing original amount by term in month
df_clean['LoanAmountOverMonth'] = df_clean['LoanOriginalAmount'] / df_clean['Term']
```

```python
fig = plt.figure(figsize=(16, 6))
ax = plt.subplot(1, 2, 1)
plt.hist(df_clean['LoanOriginalAmount'], color='skyblue',
         bins=np.logspace(np.log10(600.0), np.log10(100000.0), 20))
ax.set_title('Distribution of Loan Original Amount')
ax.set_xscale('log')
ax.set_xlabel('Amount (USD)')

ax = plt.subplot(1, 2, 2)
plt.hist(df_clean['LoanAmountOverMonth'], color='skyblue',
         bins=np.logspace(np.log10(10.0), np.log10(10000.0), 20))
ax.set_title('Distribution of Loan Monthly Repayment')
ax.set_xscale('log')
ax.set_xlabel('Amount (USD)')
plt.show()
```

As seen in the above histograms in log-scales, it appears that the distribution of original amounts are more "bumpy", indicating that total amount of loans tend to be more concentrated around certain values. On the other hand, the distribution of normalised amount is smoother, indicating a distribution closer to a log-normal distribution. This indicate that the original loan amount is likely to be influenced by the term of the loans and thus a normalised variable may be a better variable to use.

My expectation is that loans with higher adjusted amount will tend to have higher default rate, which I will explore in the next section.

#### Purpose of loans

```python
# Definition of numerical loan types from variable definitions
ListingCategories = ['0 - Not Available', '1 - Debt Consolidation', '2 - Home Improvement',
                     '3 - Business', '4 - Personal Loan', '5 - Student Use', '6 - Auto', '7- Other',
                     '8 - Baby&Adoption', '9 - Boat', '10 - Cosmetic Procedure', '11 - Engagement Ring',
                     '12 - Green Loans', '13 - Household Expenses', '14 - Large Purchases',
                     '15 - Medical/Dental', '16 - Motorcycle', '17 - RV', '18 - Taxes', '19 - Vacation',
                     '20 - Wedding Loans']
```

Based on the above, we can see that most loans are listed as Debt Consolidation, followed by Others, Business, Home Improvement, Auto, Personal Loan, Household Expenses, Student Use, and Medical/Dental. Other categories are relatively small.

#### Employment status

Next, we look at the distribution of employment statuses:

```python
fig = plt.figure()
ax = plt.subplot(1, 1, 1)
sns.countplot(x=df_clean['EmploymentStatus'], palette=sns.color_palette(n_colors=1),
              order=df_clean['EmploymentStatus'].value_counts().index)
ax.set_title('Distribution of Employment Statuses')
ax.set_xlabel('Employment Status')
plt.xticks(rotation=45)
plt.show()
```

The majority of borrowers list employment status as Employed, followed by Full-time, Self-employed, Other, Part-time, etc. However, some of these categories appear to be ambiguous, as Employed status could mean Full-time, Part-time, or other types of employments.

In order to generate a more meaningful variable to represent employment, I propose to use the EmploymentStatusDuration column to generate an "EmployementLength" variable, which represents how long a person has been in their current type of employment in months, and 0 for those who are not currently employed. For this purpose, we will consider Not available, Other, and Retired as unemployed:

```python
employed = df_clean['EmploymentStatus'].isin(['Full-time', 'Employed',
                                              'Part-time', 'Self-employed'])

df_clean['EmploymentLength'] = employed * df_clean['EmploymentStatusDuration']
```

```python
fig = plt.figure()
ax = plt.subplot(1, 1, 1)
plt.hist(df_clean['EmploymentLength'], color='skyblue', bins=40)
ax.set_title('Distribution of Employment Lengths')
ax.set_xlabel('Duration (months)')
ax.set_ylabel('Count')

plt.show()
```

We can see in the above distribution that the majority of data are at or close to 0, meaning they have been employed for a very short time or are not currently employed. The distribution tails off logarithmiclly to the right.

#### Recommendations

Next, we look at the distribution of number of recommendations:

```python
fig = plt.figure()
ax = plt.subplot(1, 1, 1)
plt.hist(df_clean['Recommendations'], color='skyblue', bins=40)
ax.set_title('Distribution of Number of Recommendations')
ax.set_xlabel('Number of recommendations')
ax.set_ylabel('Count')
ax.set_yscale('log')
plt.show()
```

As seen in the above histogram (noting that the y-axis is in log scale), the vast majority of borrowers do not have any recommendations. Others have 1-2 recommendatiosn, and only a very small amount of borrowers having 3 or more recommendations.

#### Credit score

Next, we explore the credit score distributions. It is noted that the dataset contains 2 main sources of credit scoring, the rating from Prosper and a third-party credit score. It would be interesting to compare which one performs better at predicting loan outcomes. Since both measures were likely built using similar sets of data (a borrower's credit history), they are likely highly correlated, and for the purpose of being used to predict a loan outcome statistically, only one variable should be selected.

For third-party score, we engineer a new variable by taking the average of lower and higher credit scores:

```python
df_clean['CreditScoreAvg'] = (df_clean['CreditScoreRangeUpper'] + df_clean['CreditScoreRangeLower']) / 2
```

```python
fig = plt.figure(figsize=(16, 6))
ax = plt.subplot(1, 2, 1)
sns.countplot(x=df_clean['ProsperRating (numeric)'],
              palette=sns.color_palette(n_colors=1))
ax.set_title('Distribution of Prosper Rating')
ax.set_xlabel('Prosper Rating (numeric)')

ax = plt.subplot(1, 2, 2)
plt.hist(df_clean['CreditScoreAvg'], color='skyblue', bins=19)
ax.set_title('Distribution of Average Credit Score')
ax.set_xlabel('Avg Credit Score')
plt.show()
```

For Prosper Rating, the rating of 3 (aka D) is the highest, while 7 (aka AA) is lowest. Other ratings are of relatively similar amounts.

For average credit score, the majority of borrowers are rated between 600 - 80, with a total range of around 500 - 900. Distribution appears to be right-skewed.

#### Total loan and credit amount

```python
fig = plt.figure(figsize=(16, 6))
ax = plt.subplot(1, 2, 1)
plt.hist(df_clean['OpenRevolvingMonthlyPayment'], color='skyblue', bins=10)
ax.set_title('Distribution of Open Revolving Loan Monthly Payments')
ax.set_xlabel('Amount (USD)')
ax.set_ylabel('Count')
ax.set_yscale('log')

ax = plt.subplot(1, 2, 2)
plt.hist(df_clean['AvailableBankcardCredit'], color='skyblue', bins=10)
ax.set_title('Distribution of Available Bank Card Credit')
ax.set_yscale('log')
ax.set_xlabel('Amount (USD)')
plt.show()
```

For both distributions (noting that y-axis is in log scale), we see that the majority of values are 0 or close to 0.

#### Debt-to-income ratio

Next, we look at the distribution of debt-to-income ratio:

```python
fig = plt.figure()
ax = plt.subplot(1, 1, 1)
plt.hist(df_clean['DebtToIncomeRatio'], color='skyblue', bins=50)
ax.set_title('Distribution of Debt-to-income Ratio')
ax.set_xlabel('Ratio')
ax.set_ylabel('Count')
ax.set_yscale('log')

plt.show()
```

#### Investment from friends amount

As the investment from friend amount is given in USD, we will normalise the value by engineering a investment from friend-to-total loan amount ratio variable:

```python
df_clean['InvestmentFromFriendsRatio'] = df_clean['InvestmentFromFriendsAmount'] / \
    df_clean['LoanOriginalAmount']
```

```python
fig = plt.figure()
ax = plt.subplot(1, 1, 1)
plt.hist(df_clean['InvestmentFromFriendsRatio'], color='skyblue', bins=20)
ax.set_title('Distribution of Investment from Friends Ratio')
ax.set_xlabel('Ratio')
ax.set_ylabel('Count')
ax.set_yscale('log')

plt.show()
```

We see that the majority of loans have 0 investments from friend. The distribution is generally biased towards the lower ratio, but with a peak at 1.0x, which indicates that a significant number of loans were 100% financed by friends.

Since it is difficult to tell from the chart which variable appears to be better at predicting the outcome, I will do simple regressions below to see which appears to have higher explanatory power:

```python
# Import statistical package
import statsmodels.api as sm
```

```python
# Drop rows where Prosper Rating is not available
df_prosper = df_clean.dropna(subset=['ProsperRating (numeric)'])

# Convert completed to dummy variable 1 as dependent variable
y = (df_prosper['LoanOutcome'] == 'Completed').astype(int)
# Set Prosper Rating as independent variable and add constant term
X = df_prosper['ProsperRating (numeric)']
X = sm.add_constant(X)
```

```python
# Get summary table of OLS regression for loan outcome vs Prosper Rating
mod = sm.OLS(y, X)
res = mod.fit()
print(res.summary())
```

```python
# Drop rows where credit score is not available
df_credit_score = df_clean.dropna(subset=['CreditScoreAvg'])

y = (df_credit_score['LoanOutcome'] == 'Completed').astype(int)
X = df_credit_score['CreditScoreAvg']
X = sm.add_constant(X)
```

```python
# Get summary table of OLS regression for loan outcome vs credit score
mod = sm.OLS(y, X)
res = mod.fit()
print(res.summary())
```

Based on the above, in both models, the Prosper Rating and credit score are both found to be significant. However, the R-squared results are quite small for both models. Prosper Rating appears to perform slightly better with R-squared of 0.056 compared to the credit score model's R-squared of 0.011.

However, it should be noted that it is not clear if Prosper Rating and credit score is on an interval or ordinal scale (is the difference between Prosper Score of 1-2 the same as Prosper Score of 2-3?). If they are not interval data then that could significantly affects their explanatory power in a model.

Going forward, I will use the Prosper Rating as the primary measure for borrower's credit rating, unless there is more evidence that credit score would be a better measure. To test this, I will run another test on Prosper Rating, but instead of using the numerical value of the Rating as a single variable, I will use dummy variable for each of the rating:

```python
y = (df_prosper['LoanOutcome'] == 'Completed').astype(int)
# Get dummy variables for each rating minus one
X = pd.get_dummies(df_prosper['ProsperRating (numeric)']).iloc[:, :-1]
X = sm.add_constant(X)
```

```python
mod = sm.OLS(y, X)
res = mod.fit()
print(res.summary())
```

The R-squared of this model is actually the same as the model made using a single variable, which seems to imply that there is no difference, thus Prosper Rating is likely already on an interval scale.

Going forward, I will use the Prosper Rating as the primary measure for borrower's credit rating, unless there is more evidence that credit score would be a better measure.

This could explain why the completion rate of 12-month loans are so high, although it does not explain directly why 60-month loans have a completion rate similar to 36-month loans even though borrowers are screened more strictly in terms of Prosper Rating
- The 12- and 36-month loan distributions show very clearly that those with higher Prosper Rating are more likely to successfully complete loans. However, for 60-month loans, the while the distribution shows the same result, the relationship appears to be significantly weaker

I think this could be explained by the fact that Ratings are generated at the time the loan is originated. It reflects the current state of the borrower's credit worthiness and thus could be used to predict how well the borrower's loan will perform in the short-to-medium-term. However, as the loan becomes longer, the credit score generated at the time the loan was originated almost has no explanatory power left for the loan outcome.

#### Loan outcome vs credit score - by number of recommendations

Lastly, I would like to explore interaction between Prosper Score, loan outcome and number of recommendations. Specifically, how much does number of recommendations affect loan outcome, compared to the effect from Prosper Rating?

In order to visualise this, I propose to use heatmap which shows the loan success rate at each combination of each loan's Prosper Rating and number of recommendations.

First, we generate a heatmap dataframe with Proper Ratings as rows and number of recommendations as columns. For each cell, the percentage of successfully completed loans is calculated:

```python
recc_heatmap = pd.DataFrame(index=np.arange(1, 8, dtype='int'),
                            columns=np.arange(0, 19, dtype='int'))

for index, row in recc_heatmap.iterrows():
    for num_recc in np.arange(0, recc_heatmap.shape[1]):
        completed = ((df_clean['LoanOutcome'] == 'Completed') &
                     (df_clean['Recommendations'] == num_recc) &
                     (df_clean['ProsperRating (numeric)'] == index)).sum()
        total = ((df_clean['Recommendations'] == num_recc) &
                 (df_clean['ProsperRating (numeric)'] == index)).sum()
        if total != 0:
            recc_heatmap.loc[index, num_recc] = completed / total

recc_heatmap = recc_heatmap.astype('float')
```

```python
fig = plt.figure(figsize=(16, 6))
ax = plt.subplot(1, 1, 1)
sns.heatmap(recc_heatmap, annot=True, fmt='.2f', center=0)
ax.set_xlabel('Number of Recommenations')
ax.set_ylabel('Prosper Rating')
ax.invert_yaxis()
ax.set_title(
    'Loan Success Ratio - Numbers of Recommendations vs Prosper Ratings')

plt.show()
```

As seen above, if the number of recommendations increases from 0 to 1, the success rate improves across all Prosper Rating groups. 

The increases from 1 to 2 and 2 to 3 do not generate as consistently an improvement as 0 to 1, but this could be due to the number of loans with 2 and above recommendations being so low that is causing inconsistent average results.

For loans with 4 recommendations and above, we can see that the loan success rate is 100%.

Alternately, we could simplify the above plot by reducing the categories of recommendations simply to "No recommendation" and "Has recommendation", which would show a clearer message that having recommendations significantly improves a loan's expected performance:

```python
recc_heatmap_simplify = pd.DataFrame()
recc_heatmap_simplify['No_recommendation'] = recc_heatmap[0]

for index, row in recc_heatmap_simplify.iterrows():
    completed = ((df_clean['LoanOutcome'] == 'Completed') &
                 (df_clean['Recommendations'] > 0) &
                 (df_clean['ProsperRating (numeric)'] == index)).sum()
    total = ((df_clean['Recommendations'] > 0) &
             (df_clean['ProsperRating (numeric)'] == index)).sum()
    if total != 0:
        recc_heatmap_simplify.loc[index, 'Has_recommendation'] = completed / total

recc_heatmap_simplify = recc_heatmap_simplify.astype('float')
```

```python
fig = plt.figure()
ax = plt.subplot(1, 1, 1)
sns.heatmap(recc_heatmap_simplify, annot=True, fmt='.2f', center=0)
ax.set_ylabel('Prosper Rating')
ax.invert_yaxis()
ax.set_title(
    'Loan Success Ratio - Recommendations vs Prosper Ratings')

plt.show()
```

It is also noticable that, for lower Prosper Rating, the improvement in loan outcome success is significantly more greatly improved by having recommendations than at higher Prosper Rating. The maximum improvement is at Prosper Rating of 2, with improvement of 18% added to the success rate.
